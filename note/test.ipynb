```python
learning_rate = 0.01
num_epochs = 1000
step_display = 50

x = [2.7, 4.8, 9.3, 13.4, 24.8, 31.3, 48.5, 53.0, 68.1, 74.2, 88.8, 94.5]
y = [7.0, 28.8, 22.8, 67.1, 48.8, 100.2, 140.0, 190.2, 215.2, 285.6, 260.3, 251.1]
assert len(x) == len(y)

plt.scatter(x, y)
plt.show()
```

```python
weight = tf.Variable(tf.random.uniform([1], -1.0, 1.0), name='weight')
bias = tf.Variable(tf.random.uniform([1], -1.0, 1.0), name='bias')
```

```python
y_pred = tf.add(tf.multiply(weight, x), bias)
cost = tf.reduce_mean(tf.square(y_pred - y))
```
