```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.Series( 4 * np.random.rand(4), index = list('ABCD'), name = 'Market Share')
series.plot.bar()
print(series)
plt.show()
```

```python
df = pd.DataFrame(4 * np.random.rand(5, 4), columns = list('ABCD'))
print(df)
df.plot.bar()
df.plot.barh()
df.plot.bar(stacked = True)
plt.show()
```
