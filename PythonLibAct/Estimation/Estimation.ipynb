```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Dataset
df = pd.read_csv('iris_clabel.csv')
data = np.array(df)

print(df[df['Label'] != 1])
```

```python
# Get data that have the label 1
data1 = np.array(df[df['Label'] != 1])
print(data1[0:10])
```

```python
# 
plt.hist(data1[:, 2])
plt.show()
```

```python
# Parametric Estimation
# Point estimation go go
mu = np.mean(data1, axis=0) # column vector
print(mu)
sigma = np.std(data1, axis=0) # column vector
print(sigma)
```

```python
x = np.linspace(mu[2] - 3*sigma[2], mu[2] + 3*sigma[2], 100) # Continue go go
plt.plot(x, stats.norm.pdf(x, mu[2], sigma[2]))
plt.show()
```

```python
# Nonparametric Estimation
# Kernel Density Estimation go go
estimator = stats.gaussian_kde(data1[:, 2])
K = estimator(x)
plt.plot(x, K) # x Recycle go go
plt.show()
```
