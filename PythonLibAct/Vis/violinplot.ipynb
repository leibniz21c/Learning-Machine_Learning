```python
import pandas as pd
import seaborn as sns; sns.set(style = 'whitegrid')
import matplotlib.pyplot as plt
```

```python
# Iris

df = pd.read_csv('iris_clabel.csv')
print(df)
```

```python
sns.violinplot(data = df, x = 'Label', y = 'A')
plt.title('Violin plot of Iris Dataset')
plt.show()
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.violinplot(x = 'day', y = 'tip', hue = 'sex', data = tips, split = True)
plt.show()
```
