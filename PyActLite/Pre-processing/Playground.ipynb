```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
```

```python
data = fetch_olivetti_faces()

images = data.images
print(np.shape(images)) # 400ìž¥ (64 X 64)
```

```python
x = images[0,:,:] # 0ë²ˆì§¸ ì‚¬ì§„ë§Œ ê°€ì ¸ì™€ì„œ ì €ìž¥
print(x)
plt.imshow(x)
```

```python
# min-max scaling
from sklearn import preprocessing
x_norm = preprocessing.minmax_scale(x)

print("Before Processing " + '-'*20)
print(np.min(x))
print(np.max(x))

print("\nAfter Processing " + '-'*20)
print(np.min(x_norm))
print(np.max(x_norm))

plt.imshow(x_norm)
```
