```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
print(df)
```

```python
df.plot.hist(alpha = 0.5)
plt.show()
```

```python
df.plot.hist(subplots = True, layout = (2, 2), alpha = 0.5, figsize = (8, 4))
plt.show()
```

```python
df.plot.hist(alpha = 0.5, subplots = True, layout = (2, 2), figsize = (8, 4), bins = 50)
plt.show()
```
