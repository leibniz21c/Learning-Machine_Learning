```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100, axis=0)

def threshold_func(x):
    if x >= 0 : return 1
    return 0

def sigmoid_func(x, a):
    return 1 / (1 + np.exp(- (a * x)))



y = np.array([threshold_func(t) for t in x])
plt.title('Threshold Function')
plt.plot(x, y)
plt.show()


```

```python
y = np.array([sigmoid_func(t, 3) for t in x])

plt.subplot(3, 1, 1)
plt.title('Sigmoid Function when a = 3')
plt.plot(x, y)

y = np.array([sigmoid_func(t, 10) for t in x])

plt.subplot(3, 1, 2)
plt.title('Sigmoid Function when a = 10')
plt.plot(x, y)

y = np.array([sigmoid_func(t, 100) for t in x])

plt.subplot(3, 1, 3)
plt.title('Sigmoid Function when a = 100')
plt.plot(x, y)

plt.show()
```
