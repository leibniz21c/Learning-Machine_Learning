```python
# env : local 
# 2020.09.12

# importing example
import matplotlib.pyplot as plt
```

```python
# Reading image
X = plt.imread('grey.jpg')
Y = plt.imread('five.jpg')
```

```python
# Print Image
plt.imshow(X) # X is overlapped by Y.
plt.imshow(Y)
```

```python
# Subplotting
plt.subplot(3, 3, 1) # 3 X 3 격자의 첫 번째
plt.imshow(X)
plt.subplot(3, 3, 2) # 3 X 3 격자의 두 번째
plt.imshow(5*X)
plt.subplot(3, 3, 3) # 3 X 3 격자의 세 번째
plt.imshow(255-X)

plt.subplot(3, 3, 7) # 3 X 3 격자의 일곱 번째
plt.imshow(Y)
plt.subplot(3, 3, 5) # 3 X 3 격자의 다섯 번째
plt.imshow(255-Y)
```
