```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data as DataFrame
df = pd.read_csv('iris_clabel.csv')
print(df)
```

```python
# Plot sepal-Length vs. sepal-width
df.plot.scatter(x = 'A', y = 'B')
plt.show()
# Maybe we can't use this form
```

```python
# Setting color and marker by class
colors = list('rgb')
markers = list('.*^')

fig, ax = plt.subplots(1, 1)
print(fig)
print(ax)

for i, species in enumerate(df['Label'].unique()):
    ddf = df[df['Label'] == species]
    ddf.plot.scatter(x = 'A', y = 'B',  
        alpha = 0.5, s = 10*(i+1), ax = ax, 
        color = colors[i], marker = markers[i], label = species)
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_boston
```

```python
bs = load_boston()
df = pd.DataFrame(bs.data, columns = bs.feature_names)

print(df)
```

```python
df['PRICE'] = bs.target # add price
print(df.keys())
```

```python
# Normal scatter plot
df.plot.scatter(x = 'CRIM', y = 'PRICE')
plt.title('Scatter plot of Price vs. Crime')
plt.show()
```

```python
# Log-X
df.plot.scatter(x = 'CRIM', y = 'PRICE', logx = True)
plt.title('Scatter plot of Price vs. log(Crime)')
plt.show()
```
