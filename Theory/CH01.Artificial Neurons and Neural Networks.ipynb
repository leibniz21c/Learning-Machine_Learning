1.1.2. Components<br>
- Synaptic weights
- Linear combiner
- (Nonlinear) Activation function

<img src="./res/ch01/fig_1_2.png" width="600" height="300"><br>
<p style="text-align:center">Figure.1.1.2</p>

Each $ x_1, x_2, \cdots, x_m $ are input signals, $ w_{k1}, w_{k2}, \cdots, w_{km} $ are synaptic weights. After summing junction, $ \sum_{j = 1}^{m}{w_{kj} x_j} + b_k $ is called a induced local field. Also, a bias $ b_k $ is added in this formula. So there is an affine formula. 

$$ y_k = \varphi(\sum_{j = 0}^{m}{w_{kj} x_j} + b_k) = \varphi(v_k)$$

1.1.3. Compact Representation<br>
If you suppose $ x_0 = +1 $ and $ w_{k0} = b_k $, you can write above formula compactly like this.

<img src="./res/ch01/fig_1_3.png" width="550" height="300"><br>
<p style="text-align:center">Figure.1.1.3</p>

Therefore, the formula can transform to non-affine form.<br><br>

Let vectors $ \mathbf{x} = [1, \,\ x_1, \,\ \cdots, \,\ x_m]^T $ and $ \mathbf{w} = [b, \,\ w_1, \,\ \cdots, \,\ w_m]^T $.
The induced local field $ v $ is getted this compact form.

$$ \mathbf{w}^T \mathbf{x} = \sum_i w_i x_i + b $$
Therefore, 

$$ y_k = \varphi( \sum_{j = 0}^{m} w_{wj} x_j ) = \varphi(v_k) = \varphi( \mathbf{w}_k^T \mathbf{x} ) $$

### 1.2 Activation functions
1.2.1. Threshold function(McCulloch and Pitts)

$$ 
\varphi(v) = \begin{cases} 
0 \quad if \,\ v < 0 \\
1 \quad if \,\ v \ge 0 \\
\end{cases} 
$$
$ 0 $ can be set to -1. This model follows biologically all or none law. But, this function isn't differentiable.

```python
def threshold_func(x):
    if x >= 0 : return 1
    return 0
x = np.linspace(-3, 3, 100, axis=0)
y = np.array([threshold_func(t) for t in x])

plt.title('Threshold Function')
plt.plot(x, y)
plt.show()
```

1.2.2. Sigmoid function(Logistic funtion)

$$ \varphi(v) = \frac{1}{1 + \exp(-av)} \quad for \,\ a \ge 0 $$
This model is general version of threshold function. If $ a \rightarrow \infty $, then it is threshold function. Of course, it is diffentiable. But, in flat interval, the learning doesn't work well.

```python
def sigmoid_func(x, a):
    return 1 / (1 + np.exp(- (a * x)))


x = np.linspace(-3, 3, 100, axis=0)
y = np.array([sigmoid_func(t, 3) for t in x])

plt.subplot(3, 1, 1)
plt.title('Sigmoid Function when a = 3')
plt.plot(x, y)

y = np.array([sigmoid_func(t, 10) for t in x])

plt.subplot(3, 1, 2)
plt.title('Sigmoid Function when a = 10')
plt.plot(x, y)

y = np.array([sigmoid_func(t, 100) for t in x])

plt.subplot(3, 1, 3)
plt.title('Sigmoid Function when a = 100')
plt.plot(x, y)

plt.show()
```

1.2.3. Recified Linear Unit(ReLU)<br>

$$ ReLU(x) = \max\{x, 0\} $$

It is most widely adopted and very common in the area of deep learning.


```python
def relu_func(x):
    if x > 0 :
        return x
    return 0

x = np.linspace(-3, 3, 100)
y = np.array([relu_func(t) for t in x])

plt.plot(x, y)
plt.title('ReLU Function')
plt.show()
```

1.2.4. Softplus function<br>

$$ \zeta(x) = \ln (1 + \exp(x)) $$

This is a softened version of ReLU. It is differentiable in $ x = 0 $.

```python
def softplus_func(x):
    return np.log(1 + np.exp(x))

x = np.linspace(-10, 10, 100)
y = np.array([softplus_func(t) for t in x])
plt.plot(x, y)
plt.title("Softplus function")
plt.show()
```

1.2.5. Hyperbolic tangent function<br>

$$ \tanh(x) $$

```python
x = np.linspace(-10, 10, 100)
y = np.array([np.tanh(t) for t in x])

plt.plot(x, y)
plt.title("Hyperbolic tangent function")
plt.show()
```

1.2.6. Leaky ReLU<br>

$$ f(x) = \max\{0.1x, x\} $$

This is a softened version of ReLU. It is differentiable in $ x = 0 $.

```python
def leack_relu_func(x):
    if x > 0.1*x :
        return x
    return 0.1*x 

x = np.linspace(-3, 3, 100)
y = np.array([leack_relu_func(t) for t in x])

plt.plot(x, y)
plt.title("Leack ReLU function")
plt.show()
```

1.2.7. Maxout function<br>

$$ h(x) = \max \{w_1^T x + b_1, w_2^T x + b_2\} $$

In hidden layer

1.2.8. ELU Function<br>

$$ 
ELU(x) = 
\begin{cases}
x \qquad \qquad \quad  \,\ x \ge 0 \\
\alpha (e^x - 1) \qquad x < 0 \\
\end{cases}
 $$

It sharing the property with ReLU, also the gradient doesn't be killed.

```python
def elu_function(x, alpha):
    if x >= 0 :
        return x
    return alpha*(np.exp(x) - 1)

# If alpha is zero, this is ReLU.
x = np.linspace(-3, 3, 100)
y = np.array([elu_function(t, 0) for t in x])

plt.subplot(2, 1, 1)
plt.title('ELU Function when a = 0(ReLU)')
plt.plot(x, y)

y = np.array([elu_function(t, 1) for t in x])
plt.subplot(2, 1, 2)
plt.title('ELU Function when a = 1')
plt.plot(x, y)

plt.show()
```

### 1.3 Stochastic Artificial Neuron
1.3.1. Probablity of state transition<br>

$$
x = 
\begin{cases}
+1 \quad \text{with probablity } P(v) \\
-1 \quad \text{with probablity } 1 - P(v) \\
\end{cases}
$$

1.3.2. Logistic Probability Model<br>

$$ P(v) = \frac{1}{1 + \exp(-v/T)} $$

Pseudotemperature T controls the level of synaptic noise. If $ T \rightarrow 0 $, then it is noiseless(i.e., deterministic)
