```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""

 Act01. Create another image.
 2020. 09. 12 

"""
# reading and shaping
X = plt.imread('grey.jpg')
WH = np.array(np.shape(X))

# Vectorization and normalization
x = np.reshape(X, WH[0] * WH[1]) # vectorization
x = x / 255 # normalization

# create another image
im_negative = np.reshape(1-x, WH)
im_brightening = np.reshape(3*x, WH)

# plotting
plt.subplot(2, 2, 2)
plt.imshow(np.reshape(x, WH), vmax=1, vmin=0) # unchanged
plt.subplot(2, 2, 3)
plt.imshow(im_negative, vmax=1, vmin=0) # negative image
plt.subplot(2, 2, 4)
plt.imshow(im_brightening, vmax=1, vmin=0) # brighten image
```

```python
"""

 Act02. Finding coordinates for new basis
 2020. 09. 15

"""
# current basis : Standard Orthonormal Basis
# new basis : B = {b1, b2}

# data generation
x = np.random.randn(500)
y = x + np.random.randn(500) # add correlation

# scatter plot and correlation
plt.scatter(x, y) 
print(np.corrcoef(x, y))
```

```python
# define new (orthonormal) basis
b1 = np.array([3, 4]) / np.linalg.norm([3, 4]) # Normalization
b2 = np.array([-4, 3]) / np.linalg.norm([-4, 3]) 

# data matrix
xy = np.vstack([x, y])

# calculate coordinate for each data
xy_b1 = np.zeros(500)
xy_b2 = np.zeros(500)

for i in range(500):
    xy_b1[i] = np.inner(xy[:,i], b1)
    xy_b2[i] = np.inner(xy[:,i], b2)

# scatter plot and correlation coefficient
plt.scatter(xy_b1, xy_b2)
print(np.corrcoef(xy_b1, xy_b2))
```
