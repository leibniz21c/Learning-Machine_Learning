```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.Series(3 * np.random.rand(4), index=['A', 'B', 'C', 'D'], name = "Market Share")
print(series)
series.plot.pie()
plt.show()
```

```python
df = pd.DataFrame(3 * np.random.rand(4, 1), index = list('ABCD'), columns = ['Market Share'])
df.plot.pie(subplots = True)
print(df)
plt.show()

df = pd.DataFrame(3 * np.random.rand(4, 2), index = list('ABCD'), columns = ['Market Share', 'Margin Ratio'])
df.plot.pie(subplots = True, figsize = (8, 4))
print(df)
plt.show()
 
```
