```python
# csv : Commat, Separated Value

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods=1000))
print(ts)
ts.plot()
plt.show()
```

```python
tss = ts.cumsum()
print(tss)
tss.plot()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index = pd.date_range('1/1/2000', periods=1000), columns = list('ABCD'))
print(df)
df.plot()
plt.show()
```

```python
dfs = df.cumsum()
dfs.plot()
plt.show()
```
