```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris_clabel.csv')

print(df)
```

```python
# For whole columns
df.plot.box()
plt.show()

df.boxplot(column = list('ABCD'))
plt.show()
```

```python
# For Sepal-Length(A) only
df.boxplot(column = ['A'], by = 'Label')
plt.show()
```

```python
# For all features
df.boxplot(by = 'Label')
plt.show()
```
